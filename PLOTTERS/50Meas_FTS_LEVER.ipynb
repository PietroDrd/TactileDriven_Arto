```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define the folder containing the CSV files
folder_path = "/home/rluser/thesis_ws/src/TestData_samples/var_LEVER_OH_ANNLT_8_end/"

# Initialize lists to store extracted features
Force_X_values = []
pose_x_values = []

# Iterate over each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        # Read the CSV file
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)
        
        # Extract relevant columns
        Force_X_values.append(df['Force_X'].values)
        pose_x_values.append(df['Pose_X'].values)


# Plotting
plt.figure(figsize=(10, 8))

# Subplot 1: Force_X values
plt.subplot(3, 1, 1)  # 3 rows, 1 column, subplot 1
for Force_X in Force_X_values:
    plt.plot(Force_X, label='Force_X')
plt.title('50 measurements, Switch actioning (Fx = 8N)', fontsize=16)
plt.ylabel('Force_X [N]')
plt.grid(True)

# Max and Min value calculations
rows = len(Force_X_values)
cols = len(Force_X_values[0])
maxv, minv = [], []

for i in range(cols):
    max_val, min_val = float('-inf'), float('inf')
    for j in range(rows):
        val = Force_X_values[j][i]
        if val > max_val:
            max_val = val
        if val < min_val:
            min_val = val
    maxv.append(max_val)
    minv.append(min_val)

# Difference calculations
diff = []
for a, b in zip(minv, maxv):
    diff.append(b - a)

# Subplot 2: Max and Min values
plt.subplot(3, 1, 2)  # 3 rows, 1 column, subplot 2
plt.plot(maxv, label='Max', color='blue')
plt.plot(minv, label='Min', color='orange')
plt.ylabel('Force_X [N]')
plt.grid(True)
plt.legend()

# Subplot 3: Difference between Max and Min
plt.subplot(3, 1, 3)  # 3 rows, 1 column, subplot 3
plt.plot(diff, label='Difference', color='green')
plt.ylabel('[N]')
plt.xlabel('Samples [2ms]', fontsize=14)
plt.grid(True)
plt.legend()

# Show the full figure
plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define the folder containing the CSV files
folder_path = "/home/rluser/thesis_ws/src/TestData_samples/var_LEVER_OH_ANNLT_8_end/"

# Initialize lists to store extracted features
Force_X_values = []
pose_x_values = []

# Iterate over each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        # Read the CSV file
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)
        
        # Extract relevant columns
        Force_X_values.append(df['Force_X'].values)
        pose_x_values.append(df['Pose_X'].values)

Force_X_values = pose_x_values

# Plotting
plt.figure(figsize=(10, 8))

# Subplot 1: Force_X values
plt.subplot(3, 1, 1)  # 3 rows, 1 column, subplot 1
for Force_X in Force_X_values:
    plt.plot(Force_X, label='Force_X')
plt.title('50 measurements, Switch actioning: Delta Pose X', fontsize=16)
plt.ylabel('DeltaPose_X [m]')
plt.grid(True)

# Max and Min value calculations
rows = len(Force_X_values)
cols = len(Force_X_values[0])
maxv, minv = [], []

for i in range(cols):
    max_val, min_val = float('-inf'), float('inf')
    for j in range(rows):
        val = Force_X_values[j][i]
        if val > max_val:
            max_val = val
        if val < min_val:
            min_val = val
    maxv.append(max_val)
    minv.append(min_val)

# Difference calculations
diff = []
for a, b in zip(minv, maxv):
    diff.append(b - a)

# Subplot 2: Max and Min values
plt.subplot(3, 1, 2)  # 3 rows, 1 column, subplot 2
plt.plot(maxv, label='Max', color='blue')
plt.plot(minv, label='Min', color='orange')
plt.ylabel('DeltaPose_X [m]')
plt.grid(True)
plt.legend()

# Subplot 3: Difference between Max and Min
plt.subplot(3, 1, 3)  # 3 rows, 1 column, subplot 3
plt.plot(diff, label='Difference', color='green')
plt.ylabel('[m]')
plt.xlabel('Samples [2ms]', fontsize=14)
plt.grid(True)
plt.legend()

# Show the full figure
plt.tight_layout()
plt.show()
```
